### Content Columns

*   age: age of primary beneficiary
*   sex: insurance contractor gender, female, male
*   bmi: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height,
objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9
*   children: Number of children covered by health insurance / Number of dependents
*   smoker: Smoking
*   region: the beneficiary's residential area in the US, northeast, southeast, southwest, northwest.
*   charges: Individual medical costs billed by health insurance
-------------------------------------------------

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
```

```python
df = pd.read_csv('/content/insurance (1).csv')
df.head()
```

```python
df.isnull().sum()
```

## See the relationship between each item in the dataset

```python
ax = sns.lmplot(x = 'age', y = 'charges', data=df, hue='smoker', palette='PuBu_r')
ax = sns.lmplot(x = 'bmi', y = 'charges', data=df, hue='smoker', palette='copper_r')
ax = sns.lmplot(x = 'children', y = 'charges', data=df, hue='smoker', palette='YlOrRd')
```

## Smoking is also related to the price of insurance

```python
f, ax = plt.subplots(1, 1, figsize=(10, 10))
ax = sns.violinplot(x = 'children', y = 'charges', data=df,
                 orient='v', hue='smoker', palette='PuBuGn_r')
```
